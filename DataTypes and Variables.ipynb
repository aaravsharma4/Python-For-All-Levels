<b>This Program shows why knowledge of conversion of DataTypes into other DataTypes is useful.</b>

```python
# Program to add two numbers
a = input('First number: ')
b = input('Second number: ')
print(a+b)
```

Why did this happen?
This happened as input() func takes user input. The user input is automatically converted to a string so essentially '10' + '10' = '1010'
We can use the type() func to prove this-

```python
a = input('First number: ')
b = input('Second number: ')
print(type(a))
print(type(b))
```

So python seeing they are string concatenates them together.
So the correct and useful code would be to convert it into integer and then add them.

```python
# Correct Code
# Proogram to add two numbers
a = int(input('First number: '))
b = int(input('Second number: '))
print(a+b)
```

Wouldn't it be nice if our program could also accept floats? Well we could use float which is more flexible.

```python
a = float(input('First number: '))
b = float(input('Second number: '))
print(a+b)
```
